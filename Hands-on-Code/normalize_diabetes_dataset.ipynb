```python
import numpy as np
import pandas as pd
```

## Import Data

```python
# Load data with pandas
data = pd.read_csv("diabetes.csv")
data.head()
```

```python
# Check data head
summary = data.describe()
summary
```

```python
# Check for null values
data.info()
```

```python
# Check for duplicate values
data.duplicated()
```

```python
# Normalize data
normalized_data = data.copy()

for column in normalized_data:
    if column == 'Outcome':
        continue
    # Range [0, 1)
    x = normalized_data[column]
    # a, b = -1, 1
    a, b = -np.pi, np.pi
    x = (((x - x.min()) * (b - a)) / (x.max() - x.min())) + a
    normalized_data[column] = x

normalized_data['Outcome'] = normalized_data['Outcome'].replace({0: -1, 1: 1})
normalized_data.head()
```

```python
normalized_data.to_csv('diabetes_normalized.csv', index=False)
```
